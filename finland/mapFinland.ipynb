```python
# load libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
shape = "kuntaMap/SuomenMaakuntajako_2019_10k.shp"
map_df = gpd.read_file(shape)
# check data type so we can see that this is not a normal dataframe, but a GEOdataframe
map_df.head()
```

```python
map_df.plot()
```

```python
#!pip install xlrd
# load in new csv file
df = pd.read_excel("./hoito.xlsx", header=0)
df.head()
```

```python
df=df.rename(columns={"Käynnit i": "nimi"})
print(df)
```

```python
df.drop([19,20,21])#remove rows of Ulkomaat, Tietopuuttuu, kaikki kunnat
```

```python
# join the geodataframe with the cleaned up csv dataframe
combined = map_df.set_index('NAMEFIN').join(df.set_index('nimi'))

combined.head()
```

```python
# set a variable that will call whatever column we want to visualise on the map
#variable="Kaikki vuodet"
variable = 2019

# set the range for the choropleth
vmin, vmax =44000, 3800000

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(10, 15))

# create map
combined.plot(column=variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')

# customise and add annotations

# remove the axis
ax.axis('off')

# add a title
ax.set_title(' Avohilmo: Perusterveydenhuollon asiakkaat in 2019 ', \
              
              fontdict={'fontsize': '16',
                        'fontweight' : '3'})

# create an annotation for the  data source
ax.annotate('Source:THL 2020',
           xy=(0.1, .08), xycoords='figure fraction',
           horizontalalignment='left', verticalalignment='top',
           fontsize=10, color='#555555')


# Add Labels
combined['coords'] = combined['geometry'].apply(lambda x: x.representative_point().coords[:])
combined['coords'] = [coords[0] for coords in combined['coords']]
for idx, row in combined.iterrows():
    ax.annotate(s=row['NAMESWE'], xy=row['coords'],horizontalalignment='center')

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)

# this will save the figure as a high-res png, can also save as svg
fig.savefig('testmap3.png', dpi=600)
```
